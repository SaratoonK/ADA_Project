```python
from collections import Counter
from copy import deepcopy

def real_move(tube,start,end,rep): #move from one tube to another tube
    a = tube[start].pop(-1)
    tube[end].append(a)
    if rep > 0:
        rep = rep - 1
        real_move(tube,start,end,rep)

def check_empty(tube): # check if there are empty tube
    x = None
    for i in range(len(tube)):
        if len(tube[i]) == 0:
            x = i
            return x
    return x
def top_of_tube(tube): # collect color and index of the ball at the top of each tube
    top = []
    color = []
    for i,j in enumerate(tube):
        if len(j) > 0:
            top.append((i,j[-1]))
            color.append((j[-1]))

    return top,color
def easy_move(tube,loc,depth_list): 
    # for i in range(len(loc)):
    # depth = 0
    if len(tube[loc[2]]) < 4 - depth_list[0]: # move from index1 to index 2
        real_move(tube,loc[1],loc[2],depth_list[0])
        print("case1: move normally")
        # a = tube[loc[1]].pop(-1)
        # tube[loc[2]].append(a)
    elif len(tube[loc[1]]) < 4 - depth_list[1]: # move from index2 to index1
        # a = tube[loc[2]].pop(-1)
        # tube[loc[1]].append(a)
        real_move(tube,loc[2],loc[1],depth_list[1])
        print("case2: move inversely")
    else:
        emp_index = check_empty(tube) # move to empty tube
        if emp_index is not None:
            a = tube[loc[1]].pop(-1)
            tube[emp_index].append(a)
            b = tube[loc[2]].pop(-1)
            tube[emp_index].append(b)
        print("case3: move to empty")
    print(*tube,sep = "\n")
    print("*"*50)
    temp = [tube[i][-1] for i in range(len(tube)) if len(tube[i]) != 0]
    return temp
    # print(tube)

def get_depth(tube,idx): # find out if there are stacked color or not
    depth_list = []
    # print("inside",idx)
    # idx = [0,1]
    for a in idx:
        x = 0
        for i in range(len(tube[a])-1,0,-1):
            if tube[a][i] == tube[a][i-1]:
                x = x + 1
            else:
                break
        depth_list.append(x)
    return depth_list
def check_finish(tube,idx): # check if the tube is finished
    flag = []
    # print("this is idx",idx)
    for i in range(idx):
        if len(tube[i])!= 0 and len(tube[i]) == 4:
            if tube[i][0] == tube[i][1] == tube[i][2] == tube[i][3]:
            # if tube[idx][0] == tube[idx][1] == tube[idx][2] == tube[idx][3]:
                flag.append(True)
            else:
                flag.append(False)
    return flag
        # print("hooray finish",flag)
def prep(tube):
    print("this is input",*tube,sep = "\n")
    # flag = check_finish(tube,len(tube))
    t = 0
    score = 0
    temp = ""
    while t == 0:
        if len(tube[-1]) == 0 and len(tube[-2]) == 0:
            print("first time!!")
            a = tube[0].pop(-1)
            tube[-2].append(a)
            b = tube[0].pop(-1)
            tube[-1].append(b)
            print(*tube,sep = "\n")
            print("*"*50)
        else:
            top,color = top_of_tube(tube)
            print("this is top of each tube",top)
            loc = []
            col = []
            for i in top:
                for j in top:
                    if i != j and i[1] == j[1] and i[1] not in col:
                        loc.append((i[1],i[0],j[0]))
                        col.append(i[1])
            print("this is the color to be moved from idx1 to idx 2",loc)
            # depth_list = get_depth(tube)
            for i in range(len(loc)):
                depth_list = get_depth(tube,((loc[i][1],loc[i][2])))
                # print("this is depth_list",depth_list)
                # print(tube)
                temp = easy_move(tube,loc[i],depth_list)
            # print("this is temp",temp)
            # print("this is top",top)
            if color == temp:
                print("I'm stuck")
                break
            flag = check_finish(tube,len(tube))
            if flag.count(True) == len(tube) - 2:
                score = score + 1
                break
    print("final results:")
    print(tube)
    print("*"*50)
    return score
      
```

```python
from collections import Counter

def prep_move(tube,color,dest):
    for i in range(len(dest)):
        depth_list = get_depth(tube,dest[i])
        # print("this is depth list",depth_list)
        start_idx = get_start_idx(tube,dest[i],depth_list)
        # print("this is longest tube",start_idx)
        move_condition(tube,color[i],start_idx,dest[i],depth_list)
        print(f"tube after moving {color[i]}",*tube,sep = "\n")
        print("*"*50)
    temp = [tube[i][-1] for i in range(len(tube)) if len(tube[i]) != 0]
    return temp

def get_minimum(tube,dest):
    c = 100
    for i in dest:
        if len(tube[i]) < c:
            c = len(tube[i])
            short_idx = i
    return short_idx
    
def move_condition(tube,color,start,dest,depth_list,empty = False):
    
    a = dest.index(start)
    dest.pop(a)
    print("move from tube",start)
    if len(dest) > 1:
        dest = [get_minimum(tube,dest)]
    print("move to tube",dest)
    x = depth_list.pop(a) #num_repeat for start_idx
    y = depth_list.pop(0) #num_repeat for dest_idx
    
    for i in dest:
        
        if len(tube[start]) == len(tube[i]) == 4: #only for the first time when started
            emp_index = check_empty(tube)
            if emp_index is not None:
                real_move(tube,start,emp_index,x)
                real_move(tube,i,emp_index,y)
            print("case3: move to empty")
        elif len(tube[i]) < 4 - x and tube[i][-1] == color and tube[start][-1] == color:
            real_move(tube,start,i,x)
            print("case1: move normally")
            # print(tube)
        elif len(tube[start]) < 4 - y and tube[start][-1] == color and tube[i][-1] == color:
            real_move(tube,i,start,y)
            print("case2: move inversely")
        else:
            emp_index = check_empty(tube)
            if emp_index is not None:
                a = tube[start].pop(-1)
                tube[emp_index].append(a)
                b = tube[i].pop(-1)
                tube[emp_index].append(b)
            print("case3: move to empty")
    return tube


def get_start_idx(tube,dest,depth_list):
    best = 0
    # depth_idx = 0
    # worst = 100
    # print("this is dest",dest) #[0,2]
    for k,i in enumerate(dest):
        a = len(tube[i])
        # print(a)
        if a > best:
            best = a
            long_idx = i
            depth_idx = k
            # print("this is best",best)
        elif a == best:
            if depth_list[k] > depth_list[depth_idx]:
                long_idx = i 

    return long_idx


def setup(tube):
    t = 0
    score = 0
    print("start!",*tube,sep = "\n")
    while t == 0: #start the loop
        top,_ = top_of_tube(tube) 
        print("this is top",top)
        freq = []
    
        for i in top:
            freq.append(i[1])
            
        print("this is freq",freq)
        count = dict(Counter(freq))
        freqs_sorted = sorted(count.items(), key=lambda item: item[1],reverse=True)
        print("this is freq_sorted",freqs_sorted) # count the freq of each color on top
        if freqs_sorted[0][1] == freqs_sorted[-1][1]: #check to see if the tube is done
            flag = check_finish(tube,len(tube))
            if flag.count(True) < len(tube)-2:
                a = flag.index(False)
                emp_index = check_empty(tube)
                b = tube[a].pop(-1)
                # a = tube[start].pop(-1)
                tube[emp_index].append(b)
            else:
                score = score + 1
                break
        else:
            move_color = []
            destination = []
            for i in freqs_sorted:
                if i[1] > 1:
                    move_color.append(i[0])
            print("this is move color",move_color)
            for i in move_color:
                specific = []
                for j in (top):
                    if j[1] == i:
                        specific.append(j[0])
                destination.append(specific)
            print("this is move destination",destination)
            temp = prep_move(tube,move_color,destination)
            # print("this is temp",temp)
            if freq == temp: 
                #if before/after move the top is the same. it means it's stuck
                print("Sorry, I'm stuck")
                break

    print("final results:")
    print(tube)
    print("*"*50)
    return score
```
